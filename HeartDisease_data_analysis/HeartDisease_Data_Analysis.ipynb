```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart.csv')
df.head()
```

```python
df.info()
```

```python
df['ChestPainType'].unique()
```

## Distribution plot for numerical variables

```python
sns.distplot(df['Age'],rug=True,bins=10)
```

```python
sns.distplot(df['Cholesterol'],bins=10)
```

```python
sns.distplot(df['RestingBP'],bins=10)
```

## Visualization for Categorical Variables

```python
df.groupby('Sex').size().plot(kind='pie',autopct='%.1f')
```

```python
df.groupby('ChestPainType').size().plot(kind='pie',autopct='%.1f')
```

```python
df.groupby('HeartDisease').size().plot(kind='pie',autopct='%.1f')
```

## Violin Plots

```python
sns.violinplot(data=df,x='Age')
```

```python
sns.violinplot(data=df,y='Age',x='HeartDisease')
```

```python
sns.violinplot(data=df,y='Age',x='HeartDisease',hue='Sex')
```

```python
sns.violinplot(data=df,y='Sex',x='HeartDisease')
# in this plot we can see that chances of heart disease is more in males then in females.
```

## Correlations

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

## Joint Plot


```python
sns.jointplot(data=df,x='Age',y='MaxHR')
```

```python
sns.jointplot(data=df,x='Age',y='MaxHR',kind='reg')
```

## Pair Plot

```python
sns.pairplot(df)
```
